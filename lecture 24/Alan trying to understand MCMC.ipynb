```julia
#There is a mean μ₀, that we presume has a prior distribution
 μ₀ = randn()  # This is an unknown
```

```julia
n = 100
x = randn(n).+μ₀  # generate data -- x is what is known
x̄ = mean(x)       # generate sample mean 
```

```julia
# What guess should we take?
# There is a Bayes answer
```

```julia
μᵦ =  n/(n+1)* x̄
```

```julia
#Likelihood(μ) = P(X|μ) * P(μ) 
#Likelihood(μ) = e.^sum(  -(x.-μ).^2/2 ) .* e^(-μ^2/2)
Likelihood(μ) = sum(  -(x.-μ).^2 ) + (-μ^2) #LogLikelihood
```

```julia
plot(Likelihood,-2:.01:2)
plot!([μᵦ],  [Likelihood(μᵦ )], m=:dot)
```
